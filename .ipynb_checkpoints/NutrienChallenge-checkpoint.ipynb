## Weather API Data

All weather data being pulled from Government of Canada API

Chose to include Calgary and Saskatoon stations

```python
import requests

limit = 999
calgaryClimateIdentifier = 3031109 # Calgary Springbank
saskatoonClimateIdentifier = 4057165 # Saskatoon RCS

# api-endpoint
calgaryURL = f'https://api.weather.gc.ca/collections/climate-monthly/items?f=json&lang=en-CA&limit={limit}&CLIMATE_IDENTIFIER={calgaryClimateIdentifier}'
saskatoonURL = f'https://api.weather.gc.ca/collections/climate-monthly/items?f=json&lang=en-CA&limit={limit}&CLIMATE_IDENTIFIER={saskatoonClimateIdentifier}'
  
# sending get request and saving the response as response object
rCalgary = requests.get(url = calgaryURL)
rSaskatoon = requests.get(url = saskatoonURL)
  
# extracting data in json format
calgaryData = rCalgary.json()
saskatoonData = rSaskatoon.json()
  
```

```python
print("Number of Calgary records: ", len(calgaryData['features']))
print("Number of Saskatoon records: ", len(saskatoonData['features']))
print("Example:")
calgaryData['features'][0]
```

```python
# df = pd.DataFrame.from_dict(calgaryData["features"][0]["properties"], orient="index")
# df
```

```python
# df.T
```

```python
weatherDF = pd.DataFrame()

for feature in calgaryData['features']:
    df_current = pd.DataFrame.from_dict(feature["properties"], orient="index").T
    weatherDF = pd.concat([weatherDF,df_current])
    
for feature in saskatoonData['features']:
    df_current = pd.DataFrame.from_dict(feature["properties"], orient="index").T
    weatherDF = pd.concat([weatherDF,df_current])
    
weatherDF = weatherDF.sort_values(by=['LOCAL_DATE'])
```

```python
weatherDF
```

```python
weatherDF['dateQuarter'] = pd.to_datetime(weatherDF['LOCAL_DATE'])
```

```python
quarterlyWeatherDF = weatherDF.groupby([weatherDF['dateQuarter'].dt.to_period('Q'), 'STATION_NAME'])['MEAN_TEMPERATURE'].agg('mean')
```

```python
quarterlyWeatherDF = weatherDF.groupby([weatherDF['dateQuarter'].dt.to_period('Q'), 'STATION_NAME']).agg({'MEAN_TEMPERATURE': ['mean'], 'TOTAL_PRECIPITATION': ['mean'],})
```

```python
quarterlyWeatherDF = quarterlyWeatherDF.reset_index(level=[0,1])
quarterlyWeatherDF.columns = quarterlyWeatherDF.columns.get_level_values(0)
quarterlyWeatherDF
```

## Fertilizer Shipping Data

```python
# data pulled from here: https://www150.statcan.gc.ca/t1/tbl1/en/tv.action?pid=3210003901
# Title: Fertilizer shipments to Canadian agriculture markets, by nutrient content and fertilizer year, cumulative data (x 1,000)
```

```python
fertilizerDF = pd.read_csv("Data/Fertilizer.csv")
```

```python
fertilizerDF['GEO'].unique()
```

```python
fertilizerDF.shape
```

```python
relevantGeos = ["Canada", 'Prairie provinces', 'Saskatchewan','Alberta']

fertilizerDF = fertilizerDF[fertilizerDF['GEO'].isin(relevantGeos)].reset_index(drop=True)
fertilizerDF.shape
```

```python
fertilizerDF['GEO'].unique()
```

```python
fertilizerDF = fertilizerDF.drop(['UOM', 'UOM_ID', 'SCALAR_FACTOR', 'SCALAR_ID', 'VECTOR', 'COORDINATE', 'STATUS', 'SYMBOL', 'TERMINATED', 'DECIMALS'], axis = 1)
fertilizerDF
```

```python
quarterly = []
quarterName = []
quarterCat = []

for i, period in enumerate(fertilizerDF["Period"]):
    if period == "July to September":
        quarterly.append(fertilizerDF["VALUE"].iloc[i])
        quarterName.append(fertilizerDF["REF_DATE"].iloc[i][0:4] + "Q3")
        quarterCat.append("Q3")
    if period == "July to December":
        quarterly.append(fertilizerDF["VALUE"].iloc[i]-fertilizerDF["VALUE"].iloc[i-1])
        quarterName.append(fertilizerDF["REF_DATE"].iloc[i][0:4] + "Q4")
        quarterCat.append("Q4")
    if period == "July to March":
        quarterly.append(fertilizerDF["VALUE"].iloc[i]-fertilizerDF["VALUE"].iloc[i-1])
        quarterName.append(fertilizerDF["REF_DATE"].iloc[i][5:] + "Q1")
        quarterCat.append("Q1")
    if period == "July to June":
        quarterly.append(fertilizerDF["VALUE"].iloc[i]-fertilizerDF["VALUE"].iloc[i-1])
        quarterName.append(fertilizerDF["REF_DATE"].iloc[i][5:] + "Q2")
        quarterCat.append("Q2")
    
```

```python
fertilizerDF["quarterlyVal"] = quarterly
fertilizerDF["dateQuarter"] = quarterName
fertilizerDF["quarter"] = quarterCat
fertilizerDF = fertilizerDF.drop(['REF_DATE', 'Period', 'VALUE'], axis = 1)
```

```python
fertilizerDF['dateQuarter'] = pd.to_datetime(fertilizerDF['dateQuarter']).dt.to_period('Q')
```

```python
fertilizerDF
```

## Analysis and Plots

```python
saskatoonDF = quarterlyWeatherDF[quarterlyWeatherDF["STATION_NAME"].isin(["SASKATOON RCS"])]
calgaryDF = quarterlyWeatherDF[quarterlyWeatherDF["STATION_NAME"].isin(["CALGARY SPRINGBANK A"])]
```

```python
from matplotlib.pyplot import figure

plt.style.use('classic')

fig, ax1 = plt.subplots()
plt.title("Saskatchewan quarterly nitrogen usage compared to Saskatoon precipitation")

colors = {'Q1':'red', 'Q2':'blue', 'Q3':'green', 'Q4':'yellow'}

saskNitrogenDF = fertilizerDF[fertilizerDF['Fertilizer nutrient content'].isin(["Nitrogen"])][fertilizerDF['GEO'].isin(["Saskatchewan"])]

ax1.scatter(saskNitrogenDF["dateQuarter"].astype(str), \
                    saskNitrogenDF["quarterlyVal"], \
                       c=saskNitrogenDF["quarter"].map(colors))
ax1.set_ylabel('Sask. Nitrogen Shipped Volumetric Tonnes (x1000)')

ax2 = ax1.twinx()
ax2.plot(saskatoonDF["dateQuarter"].astype(str), \
        saskatoonDF["TOTAL_PRECIPITATION"], \
         color = 'k')
ax2.set_ylabel('Quarterly Precipitation in Saskatoon (mm)')

fig.tight_layout()
fig.set_figwidth(16)
fig.set_figheight(10)
ax1.xaxis.set_major_locator(plt.MaxNLocator(15))
```

# Process Summary and Observations

For this challenge, I decided it would be interesting to see how year over year, or quarterly weather trends effect fertilizer usage. 

For weather trends I used the Government of Canada API and chose to hone in on Calgary and Saskatoon weather stations, making the assumption those would be somewhat representative of Alberta and Saskatchewan overall weather conditions.

For fertilizer usage I accessed the open source data from the Government of Canada site here: https://www150.statcan.gc.ca/t1/tbl1/en/tv.action?pid=3210003901 . I broke this data down to initially focus on Canada, Prairie provinces, Alberta, and Saskatchewan.

Once noticing the lack of available data for the chosen Calgary station, I decided to investigate just Sasatchewan usage and Saskatoon related weather data. Initially I tested both comparing temperature and precipitation to fertilizer shipments, but decided the temperature was too cyclical in nature to get effective insights. Whereas with precipitation you can directly see wet vs dry years.

I plotted all of Nitrogen, Potash, Sulphur, and Phosphate Saskatchewan shipments against Saskatoon precipitation to observe how precipitation over the last 14 years has effected the usage of each fertilizer. Red marks are Q1, blue marks Q2, green marks Q3, and yellow marks Q4. Clearly the most fertilizer used is in the spring during Q2 which makes a lot of sense during planting season. It appears the last 7 or so years have been drier than the previous in Saskatchewan. Potentially due to this it seems Potash usage is skyrocketing, while sulphur steadily climbs. Phosphate production is every so slightly increasing but could likely be related to natural growth of the agriculture industry, while nitrogen usage doesn't seem to be increasing despite the drier years.
